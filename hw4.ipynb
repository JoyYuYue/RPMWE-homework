### Data Import

```python
import statsmodels.api as sm
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
data=pd.read_csv('data.csv')
output = pd.DataFrame()
data
```

```python
#Convert the data type from object to float
columns = ['RF','Mkt-RF','SMB','HML','HML-DEV','RMW','CMA','UMD']
for i in columns:
    data[i]= data[i].str.rstrip('%').astype('float') / 100.0
data['Month']=pd.to_datetime(data['Month'])
```

### Question 2
I regress each factor with other factors and store the result in a matrix. I use the OLS fit in statsmodel.

```python
data_q1 = data[['Mkt-RF','SMB','HML','RMW','CMA','UMD']].copy()[:606]
varlist = data_q1.columns.values
result_m = np.zeros((len(varlist),len(varlist)+2))
for i in range(len(varlist)):
    y = data_q1[varlist[i]]
    x = data_q1[np.concatenate([varlist[:i],varlist[i+1:]])]
    x = sm.add_constant(x)
    model = sm.OLS(y, x).fit()
    predictions = model.predict(x)
    model.params = model.params.set_value(varlist[i], None)
    model.params = model.params.set_value('R-squared', model.rsquared)
    result_m[i][0]=model.params['const']*12
    result_m[i][1]=model.params['Mkt-RF']
    result_m[i][2]=model.params['SMB']
    result_m[i][3]=model.params['HML']
    result_m[i][4]=model.params['RMW']
    result_m[i][5]=model.params['CMA']
    result_m[i][6]=model.params['UMD']
    result_m[i][7]=model.params['R-squared']
pd.DataFrame(result_m,columns = ['Intercept','Mkt-RF','SMB','HML','RMW','CMA','UMD','R-squared'],index=['Mkt-RF','SMB','HML','RMW','CMA','UMD'])
```
